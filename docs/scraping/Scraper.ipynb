# Extracción de Artículos de El Diario

```python
from bs4 import BeautifulSoup
import requests
```

```python
# Para extraer los artículos del día
url = "https://www.eldiario.net/portal/"
response = requests.get(url)
soup = BeautifulSoup(response.content, features="html.parser")
```

```python
# Extraer todos los links que tienen labels h3
titles = soup.find_all("h3")
links = []
for title in titles:
    a = title.find("a")
    if a:
        print(a["href"])
        links.append(a["href"])
```

```python
container = soup.find("div", class_="wpb_column jeg_column vc_column_container vc_col-sm-8 jeg_main_content")
articles = container.find_all("article")
print(len(articles))
print(articles[0].text)
print(articles[5].text)
print(articles[70].text)
```

```python
article_url = "https://www.eldiario.net/portal/2023/10/10/familias-afectadas-por-la-sequia-pasan-de-27-808-a-mas-de-80-000/"
article_page = requests.get(article_url)
soup_article = BeautifulSoup(article_page.text, features="html.parser")
```

```python
title = soup_article.body.find("h1", class_="jeg_post_title")
print(title.text)
date_container = soup_article.body.find("div", class_="jeg_meta_container")
date = date_container.find("div", class_="jeg_meta_date")
print(date.text)
abstract = soup_article.body.find("div", class_="content-inner").find("p")
print(abstract.text)
```

##  de los artículos (primer intento)

```python
for i, link in enumerate(links[:10]):
    article_url = link
    try:
        article_page = requests.get(article_url)
        soup_article = BeautifulSoup(article_page.text, features="html.parser")
        title = soup_article.body.find("h1", class_="jeg_post_title")
        date_container = soup_article.body.find("div", class_="jeg_meta_container")
        date = date_container.find("div", class_="jeg_meta_date")
        paragraphs = soup_article.body.find("div", class_="content-inner").find_all("p")
        abstract = ""
        for paragraph in paragraphs:
            abstract += paragraph.text
    except:
        continue
    with open(f"article_{i}.txt", "w") as f:
        content = title.text+"\n"+date.text+"\n"+abstract
        f.write(content)
```
